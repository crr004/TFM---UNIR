**INSTALACIÓN DE DEPENDENCIAS**

## **DATASET PARA LA CLASIFICACIÓN BINARIA (Fake/No Fake)**

Análisis Exploratorio del conjunto de datos **Data_Bin_Classiffier.csv** que es el resultado de unir los conjuntos de datos de la carpeta "Data" y contiene las columnas:

* **ID**: Identificador único de cada registro (`string` de la forma `BIN_1`)
* **Title**: Título de la noticia (`string`)
* **Content**: Contenido de la noticia (`string`)
* **State**: **TARGET** que indica si una noticia es Fake o no (`boolean`)
* **Dataset**: Dataset de donde proviene esa noticia (`string`)

## **DATASET PARA LA CLASIFICACIÓN DEL TOPIC**

Análisis Exploratorio del conjunto de datos **Data_Topic_Classiffier.csv** que es el resultado de unir los conjuntos de datos de la carpeta "Data" (salvo *Fake News Detector Spanish News Edition*) y contiene las columnas:

* **ID**: Identificador único de cada registro (`string` de la forma `TOP_1`)
* **Title**: Título de la noticia (`string`)
* **Content**: Contenido de la noticia (`string`)
* **Topic**: **TARGET** que indica el campo al que pertenece la noticia (`string` discreto)
* **Dataset**: Dataset de donde proviene esa noticia (`string`)
